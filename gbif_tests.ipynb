This script will currently take a name, parse the name usings GBIF's API (maybe unnecessary), \
Generate 20 suggestions from GBIF's API, and return the list



```python
import pygbif as pg
import pandas as pd

pd.set_option('display.max_colwidth', None)
```

```python
name = 'Archispirostreptus'
```

```python

if name == '':
    print('no text, no search')
else:
    parsed = pg.species.name_parser(name)
    just_names = [name['scientificName'] for name in parsed]
    suggestions = pg.species.name_suggest(q=just_names)
print(suggestions)
```

A problem we face (optimization-wise) is that we don't need to be searching through \
or returning any species that are not found in Texas. We could certainly cross-reference \
results with our local database, which wouldn't be the worst idea, but this just feels \
clunky. \
\
PLUS, if you're only finding 20 results, and none of those are within the database, \
but the NEXT 20 results ARE in the database, you risk throwing out a search that does have \
valuable results.

```python
df = pd.DataFrame(suggestions)
filtered_df = df[(df['phylum'] == 'Arthropoda') & (df['rank'].isin(['SPECIES', 'SUBSPECIES']))]
sorted_df = filtered_df.sort_values(by='rank', key=lambda col: col != 'SPECIES')
if len(filtered_df) == 0:
    print('sorry, no matches')
else:
    print(filtered_df)
```
